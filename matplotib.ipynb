```python
import matplotlib.pyplot as plt 
```

linear graph
plt.plot() is used for linear graph
show() is used to show without text before graph

c='g' is used to give the colour to the line

```python
x=[1,2,3]
y=[4,5,6]
c='g'
plt.plot(x,y,c)
plt.show()
```

plt.plot(x,y, marker='o') is used to show the cordinate marker 

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y, marker='o')
plt.show()
```

show the different graphs 

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,marker='o')
plt.plot(y,marker='o')
plt.show()
```

Bar graph

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

changes the colour of the bar c=['r']

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r']
plt.bar(x,y, color=c)
plt.show()
```

for each bar 

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','y','g','b']
plt.bar(x,y, color=c)
plt.show()
```

showing the horizontle bar using the barh() and numpy , array 

```python
import numpy as np
```

```python
x=np.array(["A","B","C","D"])
x=np.array([1,2,3,4])
c=['r','b','y','g']
plt.barh(x,y,color=c)
plt.show()
```

to change the width and hight parameter of stick

```python
 x=np.array(["A","B","C","D"])
x=np.array([1,2,3,4])
c=['r','b','y','g']
plt.bar(x,y,color=c,width=0.1)
plt.show()
```

```python
 x=np.array(["A","B","C","D"])
x=np.array([1,2,3,4])
c=['r','b','y','g']
plt.barh(x,y,color=c,height=0.1)
plt.show()
```

scatter plot using plt.scatter(x,y)    
plt.tittle("  ") is used to show the tittle 


plt.xlable("  ") is useed to show the information of x axis


plt.ylable("  ") is useed to show the information of y axis 

```python
x=[1,2,3]
y=[4,5,6]
plt.title("scatter")
plt.xlabel("x axix")
plt.ylabel("y axis")
plt.scatter(x,y,color='red')
plt.show()
```

parameter S is used to change the size of bubble in chart

```python
x=[1,2,3]
y=[4,5,6]
plt.title("scatter")
plt.xlabel("x axix")
plt.ylabel("y axis")
plt.scatter(x,y,color='red',s=45)
plt.show()
```
